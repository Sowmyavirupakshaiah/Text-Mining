```python
def Searchasin(asin):
    url="https://www.amazon.in/dp/"+asin
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def Searchreviews(review_link):
    url="https://www.amazon.in"+review_link
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
data_asin = url.split("/dp/")[1]
data_asin =  data_asin.split("/")[0]
```

```python
link=[]
for i in range(len(data_asin)):
    response=Searchasin(data_asin[i])
    soup=BeautifulSoup(response.content)
    for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
        link.append(i['href'])
```

```python
import pandas as pd
import numpy as np 
import spacy
import string 
import seaborn as sns
from nltk import PorterStemmer
from matplotlib.pyplot import imread
from matplotlib import pyplot as plt 

from nltk.corpus import stopwords
%matplotlib inline
import re
from textblob import TextBlob
```

```python
import nltk
nltk.download('stopwords')
```

```python
def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
```

```python
STOPWORDS=stopwords.words("english")
```

```python
def clean_text(text):
    ps=PorterStemmer()
    
    text=deEmojify(text) # remove emojis
    text_cleaned="".join([x for x in text if x not in string.punctuation]) # remove punctuation
    
    text_cleaned=re.sub(' +', ' ', text_cleaned) # remove extra white spaces
    text_cleaned=text_cleaned.lower() # converting to lowercase
    tokens=text_cleaned.split(" ")
    tokens=[token for token in tokens if token not in STOPWORDS] # Taking only those words which are not stopwords
    text_cleaned=" ".join([ps.stem(token) for token in tokens])
    
    
    return text_cleaned
```

```python
df['cleaned_reviews']=df['review'].apply(lambda x:clean_text(x))
```

```python
df.sample(5)
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(height=3000, width=3000, background_color='white')
wordcloud = wordcloud.generate(' '.join(df['cleaned_reviews'].tolist()))
plt.imshow(wordcloud)
plt.title("Most common words in the reviews")
plt.axis('off')
plt.show()
```

```python
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
```

```python
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
df['Subjectivity']=pd.DataFrame(df.review.apply(lambda x: getSubjectivity(x)))
```

```python
df['Polarity']=pd.DataFrame(df.review.apply(lambda x: getPolarity(x)))
```

```python
df.head()
```

```python
def sentiment_category(score):
    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"
```

```python
df['review_category']=df['Polarity'].apply(lambda x:sentiment_category(x))
df.sample(5)
```

```python
sns.countplot(df['review_category']).set_title("Distribution of Reviews Category")
```
